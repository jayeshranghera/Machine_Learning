```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
x = np.array([1,2,3,4,5,6]).reshape(-1,1) #feature
y = np.array([1,4,9,16,25,36])    #target
```

```python
model_lin = LinearRegression()
model_lin.fit(x,y)
```

```python
y_pred = model_lin.predict(x)
```

```python
plt.scatter(x,y, color = 'blue', label = 'Actual Data')
plt.plot(x,y_pred, color = 'red', label = 'Linear Fit' )
plt.legend()
plt.show()
```

```python
# we see the data is not right fit on the red line (Linear Regression)
```

```python
model_poly = PolynomialFeatures(degree = 2)

x_poly = model_poly.fit_transform(x)
```

```python
poly_reg = LinearRegression()
poly_reg.fit(x_poly, y)
```

```python
y_pred_poly = poly_reg.predict(x_poly)

plt.scatter(x, y, color="blue", label="Actual Data")
plt.plot(x, y_pred_poly, color="green", label="Polynomial Fit (deg=2)")
plt.legend()
plt.show()
```

```python
# let's try some higher degrees 
```

```python
for d in [3, 4, 5]:
    poly = PolynomialFeatures(degree=d)
    x_poly = poly.fit_transform(x)

    poly_reg = LinearRegression()
    poly_reg.fit(x_poly, y)

    y_pred_poly = poly_reg.predict(x_poly)

    plt.scatter(x, y, color="blue")
    plt.plot(x, y_pred_poly, label=f"Degree {d}")
    plt.legend()
    plt.show()
```

# try Salary Position Dataset 

```python
import pandas as pd
data = pd.read_csv('position_salaries.csv')

x = data.iloc[:, 1:2].values    #level column 
y = data.iloc[:, 2].values      #salary column
data.head()
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x, y)

plt.scatter(x, y, color="blue")
plt.plot(x, lin_reg.predict(x), color="red", label="Linear Fit")
plt.title("Linear Regression (Salary vs Position)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
# we clearly see the red line not capture the data properly.
```

```python
poly = PolynomialFeatures(degree = 4)
x_poly = poly.fit_transform(x)
```

```python
poly_reg = LinearRegression()
poly_reg.fit(x_poly, y)
```

```python
plt.scatter(x, y, color="blue")
plt.plot(x, poly_reg.predict(x_poly), color="green", label="Polynomial Fit (deg=4)")
plt.title("Polynomial Regression (Salary vs Position)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
